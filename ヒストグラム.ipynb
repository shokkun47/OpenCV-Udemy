```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("src/Lena.jpg")
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
color_list = ["blue", "green", "red"]
for i,j in enumerate(color_list):
    hist = cv2.calcHist([img],[i], None, [256], [0,256])
    plt.plot(hist, color = j)
```
