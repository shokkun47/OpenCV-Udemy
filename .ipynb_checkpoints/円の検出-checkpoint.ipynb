```python
import cv2
import numpy as np
```

```python
img = cv2.imread("src/grapes.jpg")
img_g = cv2.imread("src/grapes.jpg", 0)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
circles = cv2.HoughCircles
```

```python
circles = cv2.HoughCircles(img_g, cv2.HOUGH_GRADIENT, dp = 1, minDist = 1, param1 = 20, param2 = 35, minRadius = 1, maxRadius = 30)
```

```python
for i in circles:
    cv2.circle(img, (i[0],i[1]),i[2], (255, 0, 0), 1)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
