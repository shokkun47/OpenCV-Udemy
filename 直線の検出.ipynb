```python
import cv2
import numpy as np
```

```python
img = cv2.imread("src/road.jpg")
img_g = cv2.imread("src/road.jpg")
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_canny = cv2.Canny(img_g, 300, 450)
cv2.imshow("img", img_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lines = cv2.HoughLines(img_canny, 1, np.pi/180, 100)
```

```python
lines
```

```python
for i in lines[:]:
    rho = i[0][0]
    theta = i[0][1]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = rho * a
    y0 = rho * b
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 1)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
